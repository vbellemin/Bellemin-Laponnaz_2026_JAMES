## Plotting the minimization trajectory 

```python
import xarray as xr
import numpy as np 
import matplotlib.pyplot as plt
```

```python
minimization_trajectory = xr.open_dataset("/silenus/PROJECTS/pr-data-ocean/bellemva/scratch/final_experiment_hawaii_itg/config_QG/minimization_trajectory.nc")
```

```python
minimization_trajectory
```

```python
plt.figure(figsize=(15,8))
plt.plot(minimization_trajectory.Nj,minimization_trajectory.J)
plt.yscale("log")   
# plt.xlim(250,350)
```

```python
traj_j = minimization_trajectory.J
f_diff = (traj_j[0:-1] - traj_j[1:])/np.maximum(np.abs(traj_j[0:-1]),np.abs(traj_j[1:]))
```

```python
f_diff[f_diff<0]=np.nan
```

```python
def stopping_criteria(arr_J,value):
    if np.all(arr_J<value):
        return True
    else: 
        return False
```

```python
for i in range(f_diff.shape[0]):
    if stopping_criteria(f_diff[i:i+20],1e-3):
        print(i)
        break
```

```python
fig, ax1 = plt.subplots(figsize=(15,8))

color = 'tab:red'
ax1.set_xlabel('Call to cost and grad function')
ax1.set_ylabel('ftol', color=color)
ax1.plot(minimization_trajectory.Nj[0:-1], f_diff, color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_yscale("log")
# ax1.set_xlim(325,350)

ax2 = ax1.twinx()  # instantiate a second Axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('gtol', color=color)  # we already handled the x-label with ax1
ax2.plot(minimization_trajectory.Ng[0:-1], minimization_trajectory.G[0:-1]/1.3e4, color=color,alpha=0.6)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_yscale("log")

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
